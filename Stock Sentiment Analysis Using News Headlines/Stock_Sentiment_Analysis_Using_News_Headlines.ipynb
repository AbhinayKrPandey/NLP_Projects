## STOCK SENTIMEANT ANALYSIS USING NEWS HEADLINES


```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Data.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
# Removing punctuations
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
##combining all the tops to one paragraph here is example for first index


' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
##iterating for all the index

headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
import nltk
```

```python
##using lemmatization and stopwords
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
nltk.download('punkt_tab')
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
for i in range(len(headlines)):
  words=nltk.word_tokenize(headlines[i])
  words=[lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
  headlines[i]=' '.join(words)
```

```python
headlines[0]
```

```python
###doing same thing for test data

data_test=test.iloc[:,2:27]
data_test.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data_test.columns= new_Index
data_test.head(5)
```

```python
# Convertng headlines to lower case for test data
for index in new_Index:
    data_test[index]=data_test[index].str.lower()
data_test.head(1)
```

```python

' '.join(str(x) for x in data_test.iloc[1,0:25])
```

```python
##iterating for all the index for test data

headlines_test = []
for row in range(0,len(data_test.index)):
    headlines_test.append(' '.join(str(x) for x in data_test.iloc[row,0:25]))
```

```python
headlines_test[0]
```

```python
for i in range(len(headlines_test)):
  words=nltk.word_tokenize(headlines_test[i])
  words=[lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
  headlines_test[i]=' '.join(words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implementing BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
test_dataset = countvector.transform(headlines_test)
predictions = randomclassifier.predict(test_dataset)
```

```python
new_var = randomclassifier.score(test_dataset,test['Label'])
new_var
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
